```python
import pandas as pd
df  = pd.read_csv('insurance.csv')
df.head(5)
```

```python
df['sex']  = df['sex'].replace({'male':1, 'female':0}).astype(int)
df['region']  = df['region'].replace({'southeast':1, 'southwest':2, 'northeast':3, 'northwest':4}).astype(int)
df['smoker']  = df['smoker'].replace({'yes':1, 'no':0}).astype(int)
df.dropna(inplace=True)
df.isna().sum()
```

```python
df.head(5)
target = df['charges']
features = df.drop('charges', axis='columns')
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
features = scaler.fit_transform(features)
features = pd.DataFrame(features)
target  = scaler.fit_transform(target.values.reshape(-1, 1))
target = pd.DataFrame(target)
features.head(5)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=1)
```

```python
import tensorflow as tf
from tensorflow import keras
model = keras.Sequential([keras.layers.Dense(100,input_shape = (6,),activation = 'relu'),
    keras.layers.Dense(50, activation='relu'),
    keras.layers.Dense(1,activation = 'linear')])
model.compile(optimizer = 'adam',loss = 'mean_squared_error',metrics = ['mean_absolute_error'])
model.fit(x_train,y_train,epochs = 50)
```

```python
predictions = model.predict(x_test)
print("prediction = " + str(predictions[0]) + "Actual = " + str(y_test.iloc[0]))

```
